```python
%%writefile app.py
import streamlit as st
import torch
from PIL import Image
import numpy as np
import os

def load_model(model_path):
    model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)
    return model

model = load_model('/content/yolov5/runs/train/yolov5_accidents3/weights/best.pt')
st.title ("Détection d'accidents avec YOLOv5s")

uploaded_file = st.file_uploader("Choisir une image", type=["jpg", "jpeg", "png"])
if uploaded_file is not None:
    image = Image.open(uploaded_file).convert('RGB')
    st.image(image, caption='Image chargée', use_column_width=True)
    results = model(image)

    st.subheader("Résultat de la détection")
    results = model(image)
    results.render() # call render() method
    st.image(Image.fromarray(results.ims[0]), caption='Résultat de la détection', use_column_width=True)

    st.subheader("Détails de la détection")
    for *box, conf, cls in results.xyxy[0]:
      st.write(f"classe: {model.names[int(cls)]}, confidence: {conf.item():.2f}")
```
